```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.analysis.diffraction.xrd import XRDCalculator
with MPRester ("pl21UGDCwMytgocW") as m:
             
    #Structure du cristal avec MP
    structure = m.get_structure_by_material_id("mp-1018059")
    #Creation fichier CIF 
    w = CifWriter(structure)
    w.write_file('mp-1018059.cif') 

    #Definition de la longueur d'ondes en angstrom
    WL = 1.54060
    #Calcul du modele XRD de la structure cristalline
    XRD = XRDCalculator(WL)
    #Calcul du diffractogramme du materiau 
    PATT = XRD.get_pattern(structure)
    #Creation d'un vecteur contenant les angles 2 theta
    ANGLE = PATT.x
    #Creation d'un vecteur contenant les intensites
    INT = PATT.y
    #Creation d'un vecteur contenant les indices hkl des differents pics
    HKL = PATT.hkls
```

# 1) Diffractogramme pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å

```python
#Affichage du diffractogramme
XRD.show_plot(structure,annotate_peaks=False,two_theta_range=[0,161])
```

# 2) 3 premiers pics du diffractogramme pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å

```python
#Affichage des trois premiers pics avec leurs indices hkl sur le diffractogramme
XRD.show_plot(structure,annotate_peaks=True,two_theta_range=[0,25])
#Affichage des resultats
print('Le premier pic possede un angle de',ANGLE[0],'avec un intensite de',INT[0],', ses indices hkl sont',HKL[0][0]['hkl'],
      'alors que la multiplicite est de',HKL[0][0]['multiplicity'])
print('Le second pic possede un angle de',ANGLE[1],'avec un intensite de',INT[1],', ses indices hkl sont',HKL[1][0]['hkl'], 
      'alors que la multiplicite est de',HKL[1][0]['multiplicity'])
print('Le toisieme pic possede un angle de',ANGLE[2],'avec un intensite de',INT[2],', ses indices hkl sont',HKL[2][0]['hkl'],
      'alors que la multiplicite est de',HKL[2][0]['multiplicity'])
```
